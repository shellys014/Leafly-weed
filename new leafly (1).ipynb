# Normalise Zip code

In Zipcode column zipcode is given in three formats so we have to convert all into single format, actual zipcode is 
of 5 digit but somewhere code is given in 4 digits so i put "0" before that to make it of 5 digit and where it is given as 9 digit i take only 5 digits 

```python
import re
import string
```

```python
df1['StoreZIP'].convert_dtypes()
```

```python
pd.to_numeric(df1['StoreZIP'], errors='coerce')
```

```python
df1['StoreZIP']=pd.to_numeric(df1['StoreZIP'], errors='coerce').convert_dtypes()
```

```python
df1.tail()
```

```python
regex=re.compile(r"(\b\d{5})")
```

```python
matches=re.findall(regex,str(df1['StoreZIP']))
```

```python
#fill "0" before in 4 digit code 

df1['StoreZIP']=df1['StoreZIP'].astype(str)
df1['StoreZIP']=df1['StoreZIP'].str.zfill(5)
```
